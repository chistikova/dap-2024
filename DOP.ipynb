```python
import numpy as np
```

```python
#Сгенерируйте массив на 100 элементов с случайными значениями,
#найдите его медиану, среднее квадратичное отклонение и дисперсию.
arr = np.random.random(100)

median = np.median(arr) #медиана
std_population = np.std(arr) #среднее квадратичное отклонение
variance_population = np.var(arr) #дисперсия

print("Массив:", arr[:100])
print("Медиана:", median)
print("Стандартное отклонение:", std_population)
print("Дисперсия:", variance_population)
```
